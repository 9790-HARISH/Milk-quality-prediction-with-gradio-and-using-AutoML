```python
import pandas as pd
import numpy as np
data=pd.read_csv("/MILK QUALITY DATASET (1).csv")
data
```

```python
data.isna().sum()
```

```python
data.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
data['Grade']=a.fit_transform(data['Grade'])
```

```python
data.head()
```

```python
x=data.iloc[:,0:7]
y=data.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb= GaussianNB()
gnb.fit(xtrain, ytrain)
```

```python
ypred = gnb.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score ,precision_score as ps,recall_score as rs,f1_score as fs,matthews_corrcoef as mcc
print("Accuracy_score:",accuracy_score (ytest,ypred))
print("Precision_score:",ps(ytest,ypred,average="weighted"))
print("F1_score:",fs(ytest,ypred,average="weighted"))
print("Matthews_corrcoef:",mcc(ytest,ypred))
```

```python
def classify(num):
  if num==0:
    return "High grade"
  elif num==1:
    return "Low grade"
  else:
    return "Medium grade"
```

```python
import gradio as gr
import numpy as np
def predict_grade(ph,temp,tas,odor,fat,turbi,col):
  input_array=np.array([[ph,temp,tas,odor,fat,turbi,col]])
  pred=gnb.predict(input_array)
  output=classify(pred)
  return[(output)]
```

```python
ph=gr.Slider(minimum=0,maximum=14,label="ph")
temp=gr.Number(label="Temperature")
tas=gr.Number(minimum=0,maximum=1,label="Taste(yes:1/no:0)")
odor=gr.Number(label="Odor(yes:1/no:0)")
fat=gr.Number(minimum=0,maximum=1,label="Fat(yes:1/no:0)")
turbi=gr.Number(minimum=0,maximum=1,label="Turbidity(yes:1/no:0)")
col=gr.Number(label="Color(240 to 255)")
op=gr.Textbox(label='Grade')
gr.Interface(predict_grade,inputs=[ph,temp,tas,odor,fat,turbi,col],outputs=op,live=True,title="MILK QUALITY PREDICTION").launch()
```
